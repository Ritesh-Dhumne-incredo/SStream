```python
import requests
import random
import json
from pyspark.sql.types import *
from datetime import datetime

cities = ["London", "New York", "Tokyo", "Mumbai", "Sydney", "Cairo"]

api_key = "c539777cbe564bcc9e850018251609"
city = random.choice(cities)
url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=no"

response = requests.get(url)
data = response.json()

fil_data = [{
    "city": data["location"]["name"],
    "temperature": float(data["current"]["temp_c"]),
    "humidity": int(data["current"]["humidity"]),
    "timestamp": datetime.now().isoformat()
}]

filename = f"data_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(f"/Volumes/workspace/stream/streaming/jsonSource/{filename}", "w") as f:
    json.dump(fil_data, f, indent=2)
```
